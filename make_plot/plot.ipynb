```python
# import libraries
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# read the data
apple_data = pd.read_csv('apple_quality.csv')

# view head of data
apple_data.head()
```

I will be answering the following questions with visualizatio

- A count plot to show the number of good and bad apple
- A barplot to the average sweetness of apple quality
- A scatter plot to show the relationship between Juciness and Ripeness

```python
sns.countplot(data=apple_data,
               x="Quality",
               )

# set plot title
plt.title("A count plot showing the number of good and bad apples")

# show plot
plt.show()
```

```python
sns.barplot(data=apple_data,
            x="Quality",
            y="Sweetness")

# Set plot title
plt.title("Average sweetness by apple quality")

# show plot
plt.show()
```

```python
sns.regplot(data=apple_data,
                x="Size",
                y="Weight")

# show plot
plt.show()
```
